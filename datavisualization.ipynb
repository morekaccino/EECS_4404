```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
fifa_path = "data/visualization/fifa.csv"
fifa_data = pd.read_csv(fifa_path, index_col="Date", parse_dates=True)
```

```python
fifa_data.head()
```

```python
fifa_data.describe()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=fifa_data)
```

```python
spotify_path = "data/visualization/spotify.csv"
spotify_data = pd.read_csv(spotify_path, index_col="Date", parse_dates=True)
```

```python
spotify_data.head()
```

```python
spotify_data.tail()
```

```python
plt.figure(figsize=(14,14))
plt.title("Daily global streams of popular songs in 2017-18")
sns.lineplot(data=spotify_data)
```

```python
plt.figure(figsize=(14,8))
plt.title("Shape of you")
plt.xlabel("Date")
sns.lineplot(data=spotify_data["Shape of You"], label="Shape of You")
sns.lineplot(data=spotify_data["Despacito"], label="Despacito")
```
